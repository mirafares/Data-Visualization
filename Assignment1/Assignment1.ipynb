## Question 1

Compare the mean Regular Unleaded Gasoline price among different cities in 2022. Plot the corresponding bar chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

#df = pd.read_csv('data/fueltypesall1991-2023.csv')

data = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') &
(df['Date'] > '2022-01-01') & (df[ 'Date'] < '2022-12-30')]
mean = data.mean()
mean
plt.title('The mean Regular Unleaded Gasoline price among different cities in 2022')
plt.xlabel("cities")
plt.ylabel( "Mean" )
mean.plot(kind='bar')
#mean_prices = df.groupby('City')['Price'].mean()
```

Plot the ECDF (normalized) of the Regular Unleaded Gasoline price in Ottawa for 2022.

```python
#import pandas as pd
#import statsmodels.api as sm
#df = pd.read_csv('data/fueltypesall1991-2023.csv')
#data = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & 
#          (df['Date'] > '2020-01-01') & (df['Date'] < '2020-12-28')] & (df['City'] == 'Ottawa')]
#x = df['Price']
import numpy as np

#data_sorted = np.sort(new_df)

# Calculate the ECDF
def ecdf(data):
    n = len(data)
    y = np.arange(1, n+1) / n
    x = np.sort(data)
    return x, y

# Plot the ECDF
x, y = ecdf(new_df['Ottawa'])
plt.plot(x, y, marker=".", linestyle="none")
plt.title("The ECDF (normalized) of the Regular Unleaded Gasoline price in Ottawa for 2022.")
plt.xlabel("Price")
plt.ylabel("ECDF")
plt.show()
```
